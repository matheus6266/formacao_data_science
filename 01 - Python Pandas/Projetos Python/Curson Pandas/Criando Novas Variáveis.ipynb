## Criando Novas Variáveis

```python
import pandas as pd
```

```python
# Importando o data frame

data = pd.read_csv("Dados/aluguel_residencial.csv", sep = ";")
```

```python
data.head(5)
```

```python
# Criando uma nova variável com o valor bruto

data["Valor Bruto"] = data["Valor"] + data["Condominio"] + data["IPTU"]
data.head(10)
```

```python
# Criando a variável com o valor do m2 com arredondamento de duas casas decimais

data["Valor m2"] = (data["Valor"] / data["Area"]).round(2)
data.head(10)
```

```python
# Criando a variável do m2 bruto

data["Valor Bruto m2"] = (data["Valor Bruto"] / data["Area"]).round(2)
data.head(5)
```

```python
# Criando uma variável que identifica se é casa ou apartamento. A função apply aplica a função em cada linha do data frame

casa = ["Casa", "Casa de Condomínio", "Casa de Vila"]
data["Tipo Agregado"] = data["Tipo"].apply(lambda x: "Casa" if x in casa else "Apartamento")
data.head(5)
```
